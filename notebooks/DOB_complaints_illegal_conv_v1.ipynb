```python
import pandas as pd
```

```python
# complaints data imported November, 2016
complaints = pd.read_csv('311_Service_Requests_from_2010_to_Present.csv')
```

```python
complaints.head()
```

```python
complaints.columns
```

```python
complaints = complaints.drop(['Cross Street 1', 'Cross Street 2', 'City', 'Agency Name', 'Street Name', 'Due Date', 'Borough', 'X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Location Type', 'Intersection Street 1', 'Intersection Street 2', 'Landmark', 'Facility Type', 'School Not Found', 'School or Citywide Complaint', 'Vehicle Type', 'Taxi Company Borough', 'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp', 'Bridge Highway Segment', 'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name', 'School Name', 'School Number', 'School Region', 'School Code', 'School Phone Number', 'School Address', 'School City', 'School State', 'School Zip', 'Park Facility Name', 'Park Borough'], axis=1)
```

```python
#create a date index to select data from 2015 only
dates = [elem for elem in complaints['Created Date']]
formatteddates = []
for i in range(0,len(dates)):
    formatteddates.append(dates[i][:-3])
```

```python
complaints['Date Index'] = formatteddates
complaints['Date Index'] = pd.to_datetime(complaints['Date Index'], format = "%m/%d/%Y %H:%M:%S")
```

```python
complaints['Incident Zip'] = complaints['Incident Zip'].fillna(0.0).astype(int)
```

```python
complaints = complaints[complaints['Incident Zip'] > 0]
```

```python
complaints = complaints.set_index('Date Index')
```

```python
#select complaints for 2015 only
complaints = complaints.loc['2015-01-01':'2016-01-01']
```

```python
print("Length of DF: " + str(len(complaints)) + "\n")
print(complaints.isnull().sum())
```

```python
#determine volume of complaints by 'Descriptor' and 'Complaint Type'
descriptor_vol = pd.DataFrame(complaints['Unique Key'].groupby([complaints['Descriptor']]).count())
descriptor_vol = descriptor_vol.rename(columns={'Unique Key': 'Count of Complaints'})
descriptor_vol = descriptor_vol.reset_index()
descriptor_vol = descriptor_vol.sort_values(by='Count of Complaints', ascending=0)
descriptor_vol.head(10)
```

```python
#determine volume of complaints by 'Complaint Type'
type_vol = pd.DataFrame(complaints['Unique Key'].groupby([complaints['Complaint Type']]).count())
type_vol = type_vol.rename(columns={'Unique Key': 'Count of Complaints'})
type_vol = type_vol.reset_index()
type_vol = type_vol.sort_values(by='Count of Complaints', ascending=0)
type_vol.head(10)
```

```python
#choosing to analyze complaints which handle illegal conversions
illegal_conv = complaints[complaints['Descriptor'] == 'Illegal Conversion Of Residential Building/Space']
print("Length of DF: " + str(len(illegal_conv)))
illegal_conv.head()
```

```python
#dependent variable dataframe with volume of illegal conversions by zip code
illegal_conv = pd.DataFrame(illegal_conv['Unique Key'].groupby([illegal_conv['Incident Zip']]).count())
illegal_conv = illegal_conv.rename(columns={'Unique Key': 'Count of Complaints'})
illegal_conv = illegal_conv.reset_index()
illegal_conv = illegal_conv.sort_values(by='Count of Complaints', ascending=0)
illegal_conv.head(10)
```

```python
illegal_conv.to_csv('DOB_count_illegal_conv_2015_ZC')
```
