### Check the assumptions for a linear model
Is the relationship between x and y linear?

```python
for i, elem in enumerate(cols[2:]):
    plt.figure(figsize=(8,8))
    plt.plot(model[elem], model['complaints'], 'r.')
    plt.title('figure ' + str(i) +  ' :\nscatter plot of ' + elem + ' and complaints to check for linearity', fontsize=12)
    if elem != 'violations_2014':
        plt.xlabel(elem + ' 2015', fontsize=12)
    else:
        plt.xlabel(elem, fontsize=12)
    plt.ylabel('volume of DOB illegal residence conversion complaints', fontsize=12)
```

Are the residuals normally distributed for each variable?

```python
for i, elem in enumerate(cols[2:]):
    results = smf.ols('complaints ~ ' + elem, data=model).fit()
    residuals = sorted(results.resid) # just in case it isn't sorted

    mu = np.mean(residuals) # mean of distribution
    sigma = np.std(residuals) # standard deviation of distribution

    num_bins = 20
    # the histogram of the data
    n, bins, patches = plt.hist(residuals, num_bins, normed=1, facecolor='blue')

    # add a 'normal best fit' line
    y = mlab.normpdf(bins, mu, sigma)
    plt.plot(bins, y, 'r--')
    plt.xlabel('distribution of residuals for ' + elem, fontsize=12)
    plt.ylabel('probability', fontsize=12)
    plt.title('histogram of ' + elem + ' residuals \nfitted normal distribution curve', fontsize=14)

    # tweak spacing to prevent clipping of ylabel
    plt.subplots_adjust(left=0.15)
    plt.show()
```

```python
e = []
for elem in cols[2:]:
    e.append(' + ' + elem)
e = ''.join(e)
```

Residual distribution for entire model?

```python
results = smf.ols('complaints ~ ' + e, data=model).fit()
residuals = sorted(results.resid) # just in case it isn't sorted

mu = np.mean(residuals) # mean of distribution
sigma = np.std(residuals) # standard deviation of distribution
 
num_bins = 20
# the histogram of the data
n, bins, patches = plt.hist(residuals, num_bins, normed=1, facecolor='blue')
 
# add a 'normal best fit' line
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('Distribution of Residuals for All Features', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.title('Histogram of Multi-Variate Residuals \nFitted Normal Distribution Curve', fontsize=14)
 
# tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.show()
```

### Test the accuracy of the model by splitting the dataset into train and test samples
Train test split with all explanatory variables

```python
# Using cross validation
R_IS=[]
R_OS=[]

n=1000
for i in range(n):
    X_train, X_test, y_train, y_test = train_test_split(model.ix[:,2:], model.ix[:,1], test_size=0.33)
    
    res=LinearRegression(fit_intercept=False)
    res.fit(X_train,y_train)
    #sum the errors 
    R_IS.append(1-((np.asarray(res.predict(X_train))-y_train)**2).sum()/((y_train-np.mean(y_train))**2).sum())                                                                     
    R_OS.append(1-((np.asarray(res.predict(X_test))-y_test)**2).sum()/((y_test-np.mean(y_test))**2).sum())
print("IS R-squared for {} times is {}".format(n,np.mean(R_IS)))
print("OS R-squared for {} times is {}".format(n,np.mean(R_OS)))
```
